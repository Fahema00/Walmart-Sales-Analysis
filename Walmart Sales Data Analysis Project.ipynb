```python
import pandas as pd 
df = pd.read_csv("Walmart_Sales.csv")
```

```python
import pandas as pd

# Convert 'Date' column to datetime format (without converting to string yet)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Sort by 'Store' and 'Date' in ascending order
df = df.sort_values(by=['Store', 'Date'], ascending=[True, True])

# After sorting, format the 'Date' column back to MM-DD-YYYY
df['Date'] = df['Date'].dt.strftime('%m-%d-%Y')

# Display the first few rows 
df.head()
```

```python
# Round Weekly Sales to the nearest 2 decimal places
df['Weekly_Sales'] = df['Weekly_Sales'].round(2)
df.head()
```

```python
# Round Temperature to the nearest whole number and convert to integer
df['Temperature'] = df['Temperature'].round(0).astype(int)
df.head()
```

```python
# Round Fuel Price to the nearest 2 decimal places
df['Fuel_Price'] = df['Fuel_Price'].round(2)
df.head()
```

```python
# Round CPI to the nearest 3 decimal places
df['CPI'] = df['CPI'].round(3)
df.head()
```

```python
# Round Unemployment to the nearest 3 decimal places
df['Unemployment'] = df['Unemployment'].round(3)

# Display the first few rows to verify the changes
df.head()
```

```python
# Drop rows with missing dates
df = df.dropna(subset=['Date'])

# Check for missing data after dropping the rows with missing dates
missing_data_after_drop = df.isnull().sum()
print("Missing data after dropping rows with missing dates:")
print(missing_data_after_drop)
```

```python
# Save the cleaned data to a CSV file
output_file_path = 'Walmart_Sales_Cleaned.csv'  # Name your output file here
df.to_csv(output_file_path, index=False)

print("File saved successfully as Walmart_Sales_Cleaned.csv")
```

```python
import pandas as pd        
import matplotlib.pyplot as plt  
import seaborn as sns        

df = pd.read_csv("Walmart_Sales_Cleaned.csv")
df.head()
```

```python
# Plot boxplot of Weekly Sales for Holiday vs Non-Holiday weeks
plt.figure(figsize=(11, 7))
sns.boxplot(x='Holiday_Flag', y='Weekly_Sales', data=df)
plt.title('Impact of Holidays on Weekly Sales')
plt.xticks([0, 1], ['Non-Holiday Weeks', 'Holiday Weeks'])
plt.ylabel('Weekly Sales')
plt.xlabel('Holiday Flag')
plt.show()
```

```python
# Group by store and compute the average unemployment rate for each store
store_unemployment = df.groupby('Store')['Unemployment'].mean().sort_values()

# Plot the stores with lowest and highest unemployment rates
plt.figure(figsize=(11, 7))
sns.barplot(x=store_unemployment.index, y=store_unemployment.values, palette="viridis")
plt.title('Average Unemployment Rate per Store')
plt.ylabel('Unemployment Rate')
plt.xlabel('Store')
plt.xticks(rotation=90)
plt.show()
```

```python
# Exclude non-numeric columns like 'Date' from the correlation calculation
numeric_df = df.select_dtypes(include='number')

# Compute the correlation matrix for numeric columns only
corr_matrix = numeric_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Walmart Sales Data')
plt.show()
```

```python
# Create a figure for two plots
plt.figure(figsize=(11, 5))

# Plot for Non-Holiday weeks (Holiday_Flag = 0)
plt.subplot(1, 2, 1)
sns.regplot(x='CPI', y='Weekly_Sales', data=df[df['Holiday_Flag'] == 0], scatter_kws={'alpha':0.3}, line_kws={"color":"blue"})
plt.title('Non-Holiday Weeks (Holiday_Flag = 0)')
plt.xlabel('CPI')
plt.ylabel('Weekly Sales')

# Plot for Holiday weeks (Holiday_Flag = 1)
plt.subplot(1, 2, 2)
sns.regplot(x='CPI', y='Weekly_Sales', data=df[df['Holiday_Flag'] == 1], scatter_kws={'alpha':0.3}, line_kws={"color":"red"})
plt.title('Holiday Weeks (Holiday_Flag = 1)')
plt.xlabel('CPI')
plt.ylabel('Weekly Sales')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create a regression plot for Fuel Price vs Weekly Sales
sns.regplot(x='Fuel_Price', y='Weekly_Sales', data=df, scatter_kws={'alpha':0.3}, line_kws={"color":"blue"})

plt.title('Relationship between Fuel Price and Weekly Sales')
plt.xlabel('Fuel Price')
plt.ylabel('Weekly Sales')

plt.show()
```
